```python
# Import the tools we need

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons, make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.cluster import KMeans
from sklearn.neural_network import MLPClassifier

from matplotlib.colors import ListedColormap
```

```python
# Create Dataset A: Moons dataset
X_moons, y_moons = make_moons(n_samples=500, noise=0.2, random_state=42)

# Create Dataset B: Linear dataset
X_linear, y_linear = make_classification(
    n_samples=500,
    n_features=2,
    n_redundant=0,
    n_informative=2,
    n_clusters_per_class=1,
    class_sep=2,
    random_state=42
)
```

```python
# Plot the Moons dataset
plt.figure()
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_moons, cmap='bwr')
plt.title("Dataset A: Nonlinear Moons Dataset")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()

# Plot the Linear dataset
plt.figure()
plt.scatter(X_linear[:, 0], X_linear[:, 1], c=y_linear, cmap='bwr')
plt.title("Dataset B: Linear Dataset")
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
# This function draws the decision boundary for a model

def plot_decision_boundary(model, X, y, title):
    
    # Create the background grid
    x_min = X[:, 0].min() - 1
    x_max = X[:, 0].max() + 1
    y_min = X[:, 1].min() - 1
    y_max = X[:, 1].max() + 1
    
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 300),
        np.linspace(y_min, y_max, 300)
    )
    
    # Make predictions for each point in the background
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Draw the background colors
    plt.figure()
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='bwr')
    
    # Draw the real data points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k')
    
    plt.title(title)
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.show()
```

```python
# Part 1: k-NN with k = 1, 3, and 5

k_values = [1, 3, 5]

# Scale the data because k-NN needs features to be on a similar scale
scaler_moons = StandardScaler()
X_moons_scaled = scaler_moons.fit_transform(X_moons)

scaler_linear = StandardScaler()
X_linear_scaled = scaler_linear.fit_transform(X_linear)

for k in k_values:
    
    # k-NN for Moons dataset
    knn_moons = KNeighborsClassifier(n_neighbors=k)
    knn_moons.fit(X_moons_scaled, y_moons)
    
    plot_decision_boundary(
        knn_moons,
        X_moons_scaled,
        y_moons,
        "k-NN Decision Boundary on Moons Dataset, k = " + str(k)
    )
    
    # k-NN for Linear dataset
    knn_linear = KNeighborsClassifier(n_neighbors=k)
    knn_linear.fit(X_linear_scaled, y_linear)
    
    plot_decision_boundary(
        knn_linear,
        X_linear_scaled,
        y_linear,
        "k-NN Decision Boundary on Linear Dataset, k = " + str(k)
    )
```

## Part 2: Decision Tree

```python
#  Decision Trees with max_depth = 2, 5, and None

depth_values = [2, 5, None]

for depth in depth_values:
    
    # Decision Tree for Moons dataset
    tree_moons = DecisionTreeClassifier(max_depth=depth, random_state=42)
    tree_moons.fit(X_moons, y_moons)
    
    plot_decision_boundary(
        tree_moons,
        X_moons,
        y_moons,
        "Decision Tree on Moons Dataset, max_depth = " + str(depth)
    )
    
    # Decision Tree for Linear dataset
    tree_linear = DecisionTreeClassifier(max_depth=depth, random_state=42)
    tree_linear.fit(X_linear, y_linear)
    
    plot_decision_boundary(
        tree_linear,
        X_linear,
        y_linear,
        "Decision Tree on Linear Dataset, max_depth = " + str(depth)
    )
```

## Part 3: K-Means clustering on the Moons dataset

```python
# ignore the real labels because clustering is unsupervised

k_cluster_values = [2, 3, 5]

for k in k_cluster_values:
    
    # Create and train the K-Means model
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(X_moons)
    
    # Get the final centroid positions
    centroids = kmeans.cluster_centers_
    
    # Plot the clusters
    plt.figure()
    plt.scatter(X_moons[:, 0], X_moons[:, 1], c=clusters, cmap='viridis', edgecolor='k')
    
    # Plot the centroids
    plt.scatter(
        centroids[:, 0],
        centroids[:, 1],
        c='red',
        marker='X',
        s=200,
        label='Centroids'
    )
    
    plt.title("K-Means Clustering on Moons Dataset, K = " + str(k))
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.legend()
    plt.show()
```

## Part 4: Neural Network with one hidden layer

```python
# test 4, 16, and 32 neurons

neuron_values = [4, 16, 32]

for neurons in neuron_values:
    
    # Neural Network for Moons dataset
    nn_moons = MLPClassifier(
        hidden_layer_sizes=(neurons,),
        activation='relu',
        max_iter=2000,
        random_state=42
    )
    
    nn_moons.fit(X_moons_scaled, y_moons)
    
    plot_decision_boundary(
        nn_moons,
        X_moons_scaled,
        y_moons,
        "Neural Network on Moons Dataset, Neurons = " + str(neurons)
    )
    
    # Neural Network for Linear dataset
    nn_linear = MLPClassifier(
        hidden_layer_sizes=(neurons,),
        activation='relu',
        max_iter=2000,
        random_state=42
    )
    
    nn_linear.fit(X_linear_scaled, y_linear)
    
    plot_decision_boundary(
        nn_linear,
        X_linear_scaled,
        y_linear,
        "Neural Network on Linear Dataset, Neurons = " + str(neurons)
    )
```
